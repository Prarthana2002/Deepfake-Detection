```python
import gradio as gr
import base64
import io
import torch
from torchvision import transforms
from PIL import Image
import numpy as np
import joblib
import os
from tensorflow.keras.models import load_model
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# -------------------- Device Setup --------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# -------------------- Deepfake Predictor --------------------
deepfake_model = load_model('/content/cnn_model.h5')

def process_and_predict_image(image):
    try:
        img = image.resize((128, 128))

        if img.mode != "RGB":
            img = img.convert("RGB")

        img_array = np.array(img) / 255.0
        img_array = np.expand_dims(img_array, axis=0)

        prediction = deepfake_model.predict(img_array)
        return 'Real' if prediction[0][0] < 0.5 else 'Fake'
    except Exception as e:
        return f"Error: {str(e)}"

# -------------------- TweetFake: AI vs Human Text Detector --------------------
MODEL_NAME = "roberta-base-openai-detector"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME)

def detect_ai_generated_text(input_text):
    inputs = tokenizer(input_text, return_tensors="pt", truncation=True, padding=True, max_length=512)

    with torch.no_grad():
        outputs = model(**inputs)
        probabilities = torch.nn.functional.softmax(outputs.logits, dim=-1)

    human_score = probabilities[0][0].item()
    ai_score = probabilities[0][1].item()

    result = "AI Generated Text" if ai_score > human_score else "Human-Written Text"

    return result, {"Human Score": round(human_score, 4), "AI Score": round(ai_score, 4)}

# -------------------- Gradio Interface --------------------
df_interface = gr.Interface(
    fn=process_and_predict_image,
    inputs=gr.Image(type="pil"),
    outputs=gr.Textbox(label="Deepfake Prediction")
)

tweet_interface = gr.Interface(
    fn=detect_ai_generated_text,
    inputs=gr.Textbox(lines=5, placeholder="Enter your text here..."),
    outputs=[gr.Textbox(label="Detection Result"), gr.JSON(label="Scores")]
)

app = gr.TabbedInterface([df_interface, tweet_interface], ["Deepfake Detector", "AI Text Detector"])

if __name__ == "__main__":
    app.launch()
```
